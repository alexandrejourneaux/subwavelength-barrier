# Definition of the problem

We consider a $\Lambda$ configuration with two ground states $|g_1\rangle$ and $|g_2\rangle$ and an excited state $|e\rangle$. Two laser beams couple the ground states to the excited state with Rabi frequencies $\Omega_c$ and $\Omega_p$, where $\Omega_c(x)$ depends on the position.

$\hat{H} = \frac{\hat{p}}{2 m} + (\omega_{g_1} - \omega_{g_2}) |g_2\rangle \langle g_2| + \omega_{g_1} |e\rangle \langle e| \\ + \Omega_c(x) \cos((\omega_{g_1} - \delta_{g_1}) t) (|g_1\rangle \langle e| + |e\rangle \langle g_1|) \\ + \Omega_p \cos((\omega_{g_2} - \delta_{g_2}) t) (|g_2\rangle \langle e| + |e\rangle \langle g_2|)$

# Born Oppenheimer Hamiltonian: existence of a dark state

In a first part, we study the Hamiltonian in the rotating wave approximation (RWA). The resulting time-independent Hamiltonian reads:

$\hat{H} = \frac{\hat{p}}{2 m} + \delta/2 (|g_2\rangle \langle g_2| - |g_1\rangle \langle g_1|) + \Delta |e\rangle \langle e| \\
+ \Omega_c(x)/2 (|g_1\rangle \langle e| + |e\rangle \langle g_1|) \\
+ \Omega_p/2 (|g_2\rangle \langle e| + |e\rangle \langle g_2|)$

We set $\hat{p} = 0$, placing ourselves in the Born-Oppenheimer approximation: we will see that this approximation is false, which will give rise to the sub-wavelength barrier.

We write the Hamiltonian in the base $|g_1\rangle$, $|g_2\rangle$, $|e\rangle$. For now we suppose the Raman resonance condition $\delta = 0$, and set a detuning $\Delta \neq 0$.

```python
def H_RWA(delta, Delta, wc, wp):
    return np.array([[-delta/2, 0, wc/2],
                     [0, delta/2, wp/2],
                     [wc/2, wp/2, Delta]])
```

### Example of the optical lattice

The simplest position dependency for $\Omega_c$ is the case of an optical lattice, where $\Omega_c(x) = \Omega_{c0} \sin(k x)$.

```python
def wc(x, k, wc0):
    return wc0 * np.sin(k*x)

def dx_wc(x, k, wc0):
    return wc0 * k * np.cos(k*x)
```

We compute the eigenstates of the Hamiltonian for different positions for $k = k_\text{rec}$ for different values of $\epsilon = \Omega_p / \Omega_{c0}$ with zero detunings $\Delta$ and $\delta$:

```python
eps = 0.1
delta = 0
Delta = 0
k = 1
x_array = np.linspace(-4, 4, 1000)
wc_array = np.array([wc(x, k, 1) for x in x_array])
wp = eps

eigvals = np.empty((len(x_array), 3), dtype=float)
eigvects = np.empty((len(x_array), 3, 3), dtype=complex)
for i, x in enumerate(x_array):
    eigvals[i,...], eigvects[i,...] = np.linalg.eigh(H_RWA(0, Delta, wc_array[i], wp))
```

```python
plt.plot(x_array, eigvals[:,0])
plt.plot(x_array, eigvals[:,1])
plt.plot(x_array, eigvals[:,2])
plt.xlabel(r"$k_{rec} x$")
plt.ylabel(r"Energie")
```

```python
x = [1,2,3,4,5]
y = x
plt.figure()
plt.plot(x_array, np.abs(wc_array), c="C3", label=r"$\Omega_c(x)$")
plt.plot(x_array, eps*np.ones((len(x_array),)), c="C0", label=r"$\Omega_p$")
plt.xlabel(r"Position $k_{rec}x$")
plt.ylabel("$\Omega / \Omega_{c0}$")
plt.legend()

#tikz_clean()
#tikz_save("omega_vs_space.tikz")
```

We can now show that the state of intermediate energy $| \Psi_2 \rangle$ is a dark state: it is decoupled from $|e\rangle$. The analytical expression of this dark state is $| \Psi_2 \rangle = \frac{1}{\overline{\Omega}} \Omega_p |g_1\rangle - \Omega_c |g_2\rangle$ and it has zero energy (for $\delta=0$) .

```python
for i, state in enumerate(["g1", "g2", "e"]):
    plt.plot(x_array, np.square(np.abs(eigvects[:, i, 1])), label=rf"$ | \phi \rangle =  | {state} \rangle $")
#plt.title(r"Projection of $\Psi_2$ on each basis state vs position")
plt.xlabel(r"Position $k_{rec}x$")
plt.ylabel(r"$\| \langle \phi | D \rangle \| ^2$")
plt.legend()

#tikz_clean()
#tikz_save("overlaps_RWA.tikz")
```

We can see that the dark state is mostly composed of $|g_2\rangle$ away from the zeros of $\Omega_c(x)$, and it "twists" to $|g_1\rangle$ on a layer of width proportional to $\epsilon$.

We denote the dark state $|D(x)\rangle$ and the lower and higher energy bright states respectively $|-(x)\rangle$ and $|+(x)\rangle$.

# Effective potential in the dark state basis, $\Delta = 0$

In the spirit of what Jendrzejewski does, we express the Hamiltonian in the basis $ \{ |\tilde{D}\rangle, |\tilde{-}\rangle, |\tilde{+}\rangle \}$ where $|\tilde{\psi}\rangle \ = | \psi (x_0) \rangle $ with $x_0$ a fixed position, applying to our hamiltonian the following unitary transformation:

$R^\dagger =  |\tilde{D}\rangle \langle D(x) | + |\tilde{-}\rangle \langle -(x) | + |\tilde{+}\rangle \langle +(x) |$

Let’s introduce the state $|B\rangle = 1/\sqrt{2} (|+\rangle + |-\rangle)$

The new Hamiltonian is then:

$H’ = \frac{(p - A)^2}{2m} + \hbar \bar{\Omega}(x) (| + \rangle \langle + | - | - \rangle \langle - |)$ where $A = i \hbar \frac{\Omega_p \partial_x \Omega_c(x)}{\bar{\Omega}(x)^2} (|\tilde{B}\rangle \langle \tilde{D} | - |\tilde{D}\rangle \langle \tilde{B} |)$

We place ourselve in the low $p$ limit, where $p^2$, $Ap$ and $pA$ can be neglected compared to $A^2$ and write the Hamiltonian in the  $ \{ \tilde{D},  \tilde{+},  \tilde{-} \}$ basis:

```python
D = np.array([[1, 0, 0]]).T
Plus = np.array([[0, 1, 0]]).T
Minus = np.array([[0, 0, 1]]).T

B = 1/np.sqrt(2) * (Plus + Minus)

A = lambda x: 1j * wp * dx_wc(x, k, 1) / (wp**2 + wc(x, k, 1)**2) * (B@D.T - D@B.T)

H = lambda x: A(x)@A(x) + (wp**2 + wc(x, k, 1)**2) * (Plus@Plus.T - Minus@Minus.T)
```

We show that the dark state is indeed an eigenstate of the low-momentum hamiltonian. And plot its energy vs position : the sub-wavelenght barrier emerges.

```python
x_array = np.linspace(-4, 4, 1000)

eigvals = np.empty((len(x_array), 3), dtype=float)
eigvects = np.empty((len(x_array), 3, 3), dtype=complex)
for i, x in enumerate(x_array):
    eigvals[i,...], eigvects[i,...] = np.linalg.eigh(H(x))
    
U_eff = eigvals[:,1]
```

```python
for i, state in enumerate(["D", "+", "-"]):
    plt.plot(x_array, np.square(np.abs(eigvects[:, i, 1])), label=rf"$ \| \langle$ {state} $| \Psi_2 \rangle \| ^2$")

plt.title(r"Projection of $\Psi_2$ on each basis state vs position")
plt.legend()
```

```python
plt.plot(x_array, eigvals[:,1], label=rf"$| ${state}$ \rangle$")
plt.title("Energy of the dark state vs position")
plt.xlabel(r"$k x$")
plt.ylabel(r"$E/E_R$")
```

The dark state remains an eigenstate of the hamiltonian, and it experiences an effective potential barrier.

# Schrödinger equation solving for $p \neq 0$ at zero detuning

We compute the eigenfunctions and eigenenergies without approximations for an atom trapped in a harmonic potential of frequency $\omega_t$ and, as previously, subjected to a uniform pump field and an optical lattice control field at Rabi frequencies $\Omega_p$ and $\Omega_c(x)$ satisfying the Raman resonance condition.

```python
def solve_harm(wt, epsilon, wp, Delta=0, delta=0, k_lattice=1, x_min=-4, x_max=4, N=500):
    """Returns a dictionnary containing : 
        -   "eigvals": the array of all eigenvalues, sorted ; 
            "eigvects": the 2d array of all eigenvectors (i.e. discretised eigenfunctions) in columns ; 
            "x_array": the array of positions on which the eigenfunctions are discretised (practical for plotting) ; 
            "eps": \epsilon (control - probe rabi frequency ratio) ;
            "wp": \Omega_p (probe rabi frequency) ; 
            "wt": \omega_t (frequency of the trap) ; 
            "U_eff": Discretized theoretical effective scalar potential (practical for plotting) ; 
            "e_overlap": array of squared scalar product between each eigenstate and the electronic excited state.
    """

    V_analytical = lambda x: 1/4 * wt**2 * x**2
    wc0 = wp/epsilon

    delta_x = (x_min - x_max)/(N-1)
    x_array = np.linspace(x_min, x_max, N)
    
    dx = 1 / (2*delta_x) * (np.diag(np.ones((N-1,), dtype=float), k=1)
                              - np.diag(np.ones((N-1,), dtype=float), k=-1))
    dx[0, :3] = [-3, 4, -1]
    dx[-1, -3:] = [1, -4, 3]
    
    p_ext = -1j * dx

    p = block_diag(p_ext, p_ext, p_ext)

    K_ext = - 1 / (delta_x**2) * (np.diag(-2*np.ones((N,), dtype=float), k=0)
                                              + np.diag(np.ones((N-1,), dtype=float), k=1)
                                              + np.diag(np.ones((N-1,), dtype=float), k=-1))
    
    K = block_diag(K_ext, K_ext, K_ext)
    
    zeros =  np.zeros((N, N), dtype = complex)
    identity = np.eye(N, dtype = complex)

    g1_g1 = block_diag(identity, zeros, zeros)
    g2_g2 = block_diag(zeros, identity, zeros)
    e_e = block_diag(zeros, zeros, identity)
    
    wc_array = np.array([wc(x, k_lattice, wc0) for x in x_array])
    wc_diag = np.diag(wc_array)
    
    U_eff = wp**2 * (dx@wc_array)**2 / (wp**2 + wc_array**2)**2

    wc_over2_g1_e = np.block([[zeros, zeros, wc_diag/2],
                              [zeros, zeros, zeros],
                              [zeros, zeros, zeros]])

    wc_over2_e_g1 = wc_over2_g1_e.T

    g2_e = np.block([[zeros, zeros, zeros],
                     [zeros, zeros, identity],
                     [zeros, zeros, zeros]])

    e_g2 = g2_e.T

    U = delta/2 * (g2_g2 - g1_g1) + Delta * e_e + (wc_over2_g1_e + wc_over2_e_g1) + wp/2 * (g2_e + e_g2)

    V_ext = np.diag([V_analytical(x) for x in x_array])
    V = block_diag(V_ext, V_ext, V_ext)

    H = K + V + U
    
    res = eigh(H)
    
    return {"eigvals": res[0], 
            "eigvects": res[1], 
            "x_array": x_array, 
            "eps": epsilon, 
            "wp": wp, 
            "wt": wt, 
            "U_eff": U_eff, 
            "e_overlap": np.array([np.sum(np.square(np.abs(e_e@res[1][:,i]))) for i in range(3*N)])
           }
```

```python
wt = hbar / E_recRed * 2*np.pi*10e3 #Hz
eps = 0.1
wp = 3e3
V_analytical = lambda x: 1/4 * wt**2 * x**2 # The 1/4 instead of 1/2 is due to the nondimensionalization of the mass (because a "2m" appears in E_rec = \hbar^2 k^2 / 2m)
```

```python
sol = solve_harm(wt, eps, wp)
```

We plot the normalized effective scalar potential vs $\epsilon$ : width is proportional to $\epsilon$ and height to $\epsilon^2$.

```python
for epsilon in [0.01, 0.05, 0.1, 0.5]:
    N=5000
    x_min, x_max = -4, 4
    k_lattice=1
    wc0 = wp/epsilon
    delta_x = (x_min - x_max)/(N-1)
    x_array = np.linspace(x_min, x_max, N)
    wc_array = np.array([wc(x, k_lattice, wc0) for x in x_array])
    dx = 1 / (2*delta_x) * (np.diag(np.ones((N-1,), dtype=float), k=1)
                            - np.diag(np.ones((N-1,), dtype=float), k=-1))
    dx[0, :3] = [-3, 4, -1]
    dx[-1, -3:] = [1, -4, 3]
    U_eff = wp**2 * (dx@wc_array)**2 / (wp**2 + wc_array**2)**2 * epsilon**2
    
    plt.plot(x_array, U_eff, label=rf"$\varepsilon$ = {epsilon}")
    plt.xlabel(r"Position $k_{rec}x$")
    plt.ylabel(r"Effective potential $\varepsilon^2 U(x)$")
    plt.legend()
    #tikz_clean()
    #tikz_save('effective_potential.tikz')
    
```

We plot the overlap with the excited state for the energy-sorted eigenstates. For low energies, we see the negative-enegy $| - \rangle$ state that has 1/2 overlap with $| e \rangle$, and for high energies, we see the $| + \rangle$ state that has 1/2 overlap with $| e \rangle$. In beween must be the dark state we are looking for.

# Non-zero detuning $\delta$

# Introduction of a second excited state

### Back to Born-Oppenheimer approximation 

The hamiltonian in the RWA in the presence of a second excited state reads:

$\hat{H} = \frac{\hat{p}}{2 m} + \delta/2 (|g_2\rangle \langle g_2| - |g_1\rangle \langle g_1|) + \Delta |e_1\rangle \langle e_1| + (\Delta + \delta_3 - \delta_2) | e_2 \rangle \langle e_2 | \\
+ \Omega_c(x)/2 (|g_1\rangle \langle e_1| + |e_1\rangle \langle g_1| + |g_1\rangle \langle e_2| + |e_2\rangle \langle g_1|) \\
+ \Omega_p/2 (|g_2\rangle \langle e_1| + |e_1\rangle \langle g_2|)$

```python
def H_RWA(delta, Delta, delta32, wc, wp):
    return np.array([[-delta/2, 0,       wc/2,  wc/2],
                     [0,        delta/2, wp/2,  0],
                     [wc/2,     wp/2,    Delta, 0],
                     [wc/2,     0,       0,     Delta+delta32]])
```

```python
def wc(x, k, wc0):
    return wc0 * np.sin(k*x)

def dx_wc(x, k, wc0):
    return wc0 * k * np.cos(k*x)
```

```python
eps = 0.1
delta = 0
Delta = 0
k = 1
x_array = np.linspace(-4, 4, 1000)
wc_array = np.array([wc(x, k, 1) for x in x_array])
wp = eps

n_delta32 = 7
delta32_array = np.linspace(-2, 2, n_delta32)

eigvals = np.empty((len(x_array), n_delta32, 4), dtype=float)
eigvects = np.empty((len(x_array), n_delta32, 4, 4), dtype=complex)
for i, x in enumerate(x_array):
    for j, delta32 in enumerate(delta32_array):
        eigvals[i, j,...], eigvects[i, j,...] = np.linalg.eigh(H_RWA(0, Delta, delta32, wc_array[i], wp))
```

We plot the energies vs position in the Born Oppenheimer approximation. for different $\Delta’ = \delta_3 - \delta_2$

### Time-independent Schrödinger equation in position space for the 4-states system

```python
def solve_harm4(wt, epsilon, wp, Delta=0, delta=0, delta32=0, k_lattice=1, x_min=-4, x_max=4, N=500):    
    """Returns a dictionnary containing : 
        -   "eigvals": the array of all eigenvalues, sorted ; 
            "eigvects": the 2d array of all eigenvectors (i.e. discretised eigenfunctions) in columns ; 
            "x_array": the array of positions on which the eigenfunctions are discretised (practical for plotting) ; 
            "eps": \epsilon (control - probe rabi frequency ratio) ;
            "wp": \Omega_p (probe rabi frequency) ; 
            "wt": \omega_t (frequency of the trap) ; 
            "U_eff": Discretized theoretical effective scalar potential (practical for plotting) ; 
            "e1_overlap": array of squared scalar product between each eigenstate and the first electronic excited state.
            "e2_overlap": array of squared scalar product between each eigenstate and the second electronic excited state.
    """
    
    V_analytical = lambda x: 1/4 * wt**2 * x**2
    wc0 = wp/eps

    delta_x = (x_min - x_max)/(N-1)
    x_array = np.linspace(x_min, x_max, N)
    
    dx = 1 / (2*delta_x) * (np.diag(np.ones((N-1,), dtype=float), k=1)
                              - np.diag(np.ones((N-1,), dtype=float), k=-1))
    dx[0, :3] = [-3, 4, -1]
    dx[-1, -3:] = [1, -4, 3]
    
    p_ext = -1j * dx

    p = block_diag(p_ext, p_ext, p_ext, p_ext)

    K_ext = - 1 / (delta_x**2) * (np.diag(-2*np.ones((N,), dtype=float), k=0)
                                              + np.diag(np.ones((N-1,), dtype=float), k=1)
                                              + np.diag(np.ones((N-1,), dtype=float), k=-1))

    K = block_diag(K_ext, K_ext, K_ext, K_ext)
    
    zeros =  np.zeros((N, N), dtype = complex)
    identity = np.eye(N, dtype = complex)

    g1_g1 = block_diag(identity, zeros, zeros, zeros)
    g2_g2 = block_diag(zeros, identity, zeros, zeros)
    e1_e1 = block_diag(zeros, zeros, identity, zeros)
    e2_e2 = block_diag(zeros, zeros, zeros, identity)
    
    wc_array = np.array([wc(x, k_lattice, wc0) for x in x_array])
    wc_diag = np.diag(wc_array)
    
    U_eff = wp**2 * (dx@wc_array)**2 / (wp**2 + wc_array**2)**2

    wc_over2_g1_e1 = np.block([[zeros, zeros, wc_diag/2, zeros],
                              [zeros, zeros, zeros, zeros],
                              [zeros, zeros, zeros, zeros],
                              [zeros, zeros, zeros, zeros]])

    wc_over2_e1_g1 = wc_over2_g1_e1.T
    
    wc_over2_g2_e2 = np.block([[zeros, zeros, zeros, zeros],
                              [zeros, zeros, zeros, wc_diag/2],
                              [zeros, zeros, zeros, zeros],
                              [zeros, zeros, zeros, zeros]])

    wc_over2_e2_g2 = wc_over2_g2_e2.T


    g2_e1 = np.block([[zeros, zeros, zeros, zeros],
                     [zeros, zeros, identity, zeros],
                     [zeros, zeros, zeros, zeros],
                     [zeros, zeros, zeros, zeros]])

    e1_g2 = g2_e1.T

    U = delta/2 * (g2_g2 - g1_g1) + Delta * e1_e1 + (Delta + delta32) * e2_e2  + (wc_over2_g1_e1 + wc_over2_e1_g1 + wc_over2_g2_e2 + wc_over2_e2_g2) + wp/2 * (g2_e1 + e1_g2)

    V_ext = np.diag([V_analytical(x) for x in x_array])
    V = block_diag(V_ext, V_ext, V_ext, V_ext)

    H = K + V + U
    
    res = eigh(H)
    
    return {"eigvals": res[0],
            "eigvects": res[1],
            "x_array": x_array,
            "eps": epsilon,
            "wp": wp,
            "wt": wt,
            "U_eff": U_eff,
            "e1_overlap": np.array([np.sum(np.square(np.abs(e1_e1@res[1][:,i]))) for i in range(4*N)]),
            "e2_overlap": np.array([np.sum(np.square(np.abs(e2_e2@res[1][:,i]))) for i in range(4*N)])
           }
```

```python
wt = hbar / E_recRed * 2*np.pi*10e3 #Hz
eps = 0.1
wp = 30
V_analytical = lambda x: 1/4 * wt**2 * x**2
```

```python
for delta32 in tqdm(np.logspace(1, 7, 7)):
    sol = solve_harm4(wt, eps, wp, delta32=delta32)    
    N = len(sol["x_array"])
    
    plt.figure()
    plt.plot(sol["x_array"], np.array([V_analytical(x) for x in sol["x_array"]]) + sol["U_eff"])
    for i in np.argsort(sol["e1_overlap"] + sol["e2_overlap"])[:10]:
        plt.plot(sol["x_array"], 
                 1000*(np.square(np.abs(sol["eigvects"][:N,i]))
                      +np.square(np.abs(sol["eigvects"][N:2*N,i]))
                      +np.square(np.abs(sol["eigvects"][2*N:3*N,i]))
                      +np.square(np.abs(sol["eigvects"][3*N:,i]))) 
                 + sol["eigvals"][i], label=rf"""$ \| \langle$ e1 $| \Psi \rangle \| ^2 =$ {f"{sol['e1_overlap'][i]:.2e}"}, $ \| \langle$ e2 $| \Psi \rangle \| ^2 =$ {f"{sol['e2_overlap'][i]:.2e}"}""")
    
    plt.ylim(-10, 135)
    #plt.legend()
    #plt.yscale("log")
    plt.title(rf"Darkest states probability density vs position at $\delta3 - \delta2$ = {f'{delta32:.2e}'}")
```

We want to see the impact of the coupling to the fourth state to the wavefunction of the dark state. For different $\Omega_p$, we look at the influence of $\Delta’$ on the projection of the darkest state on the excited state. We observe a quadratic dependance on $\Omega_p$ of the minimal $\Delta’$ allowing to get a dark state.

```python
wt = hbar / E_recRed * 2*np.pi*10e3 #Hz
eps = 0.1
wp = 1e3
V_analytical = lambda x: 1/4 * wt**2 * x**2
```

```python
delta32_array = np.logspace(0, 8, 50)
wp_array = np.logspace(0, 3, 20)

overlap = np.empty((len(wp_array), len(delta32_array)))

for i, wp in enumerate(wp_array):
    sol3states = solve_harm(wt, eps, wp)
    idx3states = np.argmin(sol3states["e_overlap"])
    dark_function = sol3states["eigvects"][:,idx3states]
    for j, delta32 in tqdm(enumerate(delta32_array)):
        sol = solve_harm4(wt, eps, wp, delta32=delta32)  
        N = len(sol["x_array"])
        idx = np.argmin(sol["e1_overlap"])
        overlap[i, j] = np.square(np.abs(np.conj(dark_function)@sol["eigvects"][:3*N, idx]))
```
